## Voici un algorithme qui inverse le noir et le blanc à partir d'un fichier pbm et qui enregistre les modifications dans le fichier negatif.pbm

```python
fic=open('SNT.pbm','r')     #ouverture du fichier source
neg=open('negatif.pbn','w') #création du fichier cible
for i in range(4):          #recopie des 4 premières lignes
    chaine=fic.readline()
    neg.write(chaine)

for ligne in fic:           #lecture ligne à ligne
    mod=''                  #chaine vide
    for c in ligne:         #lecture de chaque caractère de chaque ligne
        if c=='0':          #Recherche d'un pixel blanc
            mod = mod +'1 '
        if c=='1':          #Recherche d'un pixel noir
            mod = mod +'0 '
    print(mod)
    neg.write(mod+'\n')     #affichage des lignes du fichier
              
fic.close()
neg.close()
```
